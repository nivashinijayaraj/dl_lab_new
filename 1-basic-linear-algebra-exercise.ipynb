```python
import pandas as pd

import numpy as np
```

```python
#Matrix Multiplication

a=np.array([[1,2],[3,4]])

b=np.array([[5,6],[7,8]])

c=np.dot(a,b)

print(c)
```

```python
#Eigen Decomposition

A=np.array([[4,2],[1,3]])

eigenvalues,eigenvector=np.linalg.eig(A)

eigenvalues,eigenvector
```

```python
#Singular Value Decomposition

A = np.array([[3, 2, 2], [2, 3, -2]])

U, S, V = np.linalg.svd(A)

U, S, V
```

```python
#Determinant Calculation

A = np.array([[6, 1], [2, 3]])

det=np.linalg.det(A)

det
```

```python
#Norm Calculation

v=np.array([3,4])

l1=np.linalg.norm(v,ord=1)

l2=np.linalg.norm(v)

print(v)

print(l1)

print(l2)
```
